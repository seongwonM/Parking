## 인구별

```python
df1 = pd.read_excel('전주시data/전주시인구.xlsx')
df1.columns=df1.iloc[1]
df1=df1.iloc[2:]
```

```python
df1=df1.iloc[:,2:]
df1.head()
```

```python
df2 = gpd.read_file('전주시data/DA_EMD_202307.shp', encoding='cp949')
df2.head()
df3=df2.to_crs(epsg=3857)
df3=df3.iloc[:,1:]
df3.columns=['동이름','code','geometry']
df3['center']=df3.geometry.centroid
```

```python
df3
```

```python
df3['population'] = df3['population'].str.replace(',', '').astype(float)
```

```python
import matplotlib.pyplot as plt
import geopandas as gpd
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# 플롯 생성
fig = plt.figure(figsize=(18, 13))
ax = fig.subplots(1, 1)

# 범례용 구분자 추가
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)

# 데이터 플롯
df3.plot(column="population", ax=ax, legend=True, cax=cax, cmap='Blues')

# 제목 설정
ax.set_title("전주시 동별 인구 수", fontsize=25, pad=20)

# 플롯 표시 
plt.show()
```

```python
import matplotlib.pyplot as plt
import geopandas as gpd
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# 플롯 생성
fig = plt.figure(figsize=(18, 13))
ax = fig.subplots(1, 1)

# 범례용 구분자 추가
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)

# 데이터 플롯
df3.plot(column="population", ax=ax, legend=True, cax=cax, cmap='Blues')

# 제목 설정
ax.set_title("전주시 동별 인구 수", fontsize=25, pad=20)

# 지역명 추가 
for idx, row in df3.iterrows():
    # 중심 좌표 가져오기
    lat = row.geometry.centroid.y
    long = row.geometry.centroid.x

    # 인구 수에 따른 글자 색상 선택
    if row["population"] >= 400000:
        c = "white"
    else: 
        c = 'black'

    # 지역명 추가
    ax.text(long, lat, row["동이름"], fontsize=8, color=c, fontweight='bold', ha='center', va='center')

# 축 숨기기
ax.set_axis_off()

# 플롯 표시
plt.show()
```
